```python
img = imread('myImage.bmp')
img_g = color.rgb2gray(img)
pl.subplot(121), pl.imshow(img), pl.axis('off')
pl.subplot(122), pl.imshow(img_g, cmap='gray'), pl.axis('off')
pl.show()
```

```python
img = imread('myImage.bmp')

pl.figure(figsize=(10,8))

pl.subplot(131), pl.imshow(img[...,0], cmap='Reds')
pl.title('r', size=20), pl.axis('off')

pl.subplot(132), pl.imshow(img[...,1], cmap='Greens'),
pl.title('g', size=20), pl.axis('off')

pl.subplot(133), pl.imshow(img[...,2], cmap='Blues'),
pl.title('b', size=20), pl.axis('off')

pl.show()
```

```python
img = imread('myImage.bmp')
img_hsv = color.rgb2hsv(img)

pl.gray()

pl.figure(figsize=(10,8))

pl.subplot(221), pl.imshow(img_hsv[...,0])
pl.title('h', size=20), pl.axis('off')

pl.subplot(222), pl.imshow(img_hsv[...,1]),
pl.title('s', size=20), pl.axis('off')

pl.subplot(223), pl.imshow(img_hsv[...,2]),
pl.title('v', size=20), pl.axis('off')

pl.show()
```
