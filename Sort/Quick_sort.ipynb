```python
import time
import random

def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    left = quick_sort(left)
    right = quick_sort(right)

    return left + middle + right
```

```python
arr = [x for x in range(1001)] # 10000개의 원소를 갖는 리스트
random.shuffle(arr) # 정렬하기 전 무작위로 섞기
suffled_arr = arr[:] # 퀵 정렬에 사용할 리스트

start = time.time()
sorted_arr = quick_sort(suffled_arr)
end = time.time()
print("Before array: ", suffled_arr)
print("Sorted array:", sorted_arr)
print("Time sort:", end - start)
```
